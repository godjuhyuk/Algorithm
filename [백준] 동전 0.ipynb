```python
import sys
input = sys.stdin.readline
# N, K = map(int, input().split())
# coins = [x.strip() for x in sys.stdin.readlines()]
N, K = 10, 4790
coins = [1, 5, 10, 50, 100, 500, 1000, 5000, 10000, 50000]
# 그리디
# idx, ans = N-1, 0
# while K != 0:
#     val = coins[idx]
#     if val <= K:
#         ans += K // val
#         K -= val * (K // val)
#     else:
#         idx -= 1
# print(ans)

#DP
# N, K = map(int, input().split())
# coins = [int(input()) for _ in range(N)]

dp = [100000000] * (K + 1)  # dp[i]: 가치의 합이 i일 때 필요한 동전의 최소 개수
dp[0] = 0  # 가치의 합이 0인 경우 필요한 동전 개수는 0

for coin in coins:
    for i in range(coin, K + 1):
        dp[i] = min(dp[i], dp[i - coin] + 1)

dp[4000]
```
