```python
# 1차 시도 - 시간초과 
N, M = map(int, input().split())
number_arr = [x for x in range(1, N+1)]

def combinations(list, n):
    if n == 0:
        return [[]]
    
    result = []
    for i in range(len(list)):
        first_element = list[i]
        for combination in combinations(list, n-1):
            if first_element not in combination:
                result.append([first_element] + combination)
    return result

for combination in combinations(number_arr, M):
    print(*combination)
```

```python
N, M = map(int, input().split())
output = ''
state = [False] * (N+1)
pool = []
def search(deps):
    global output
    if deps >= M:
        # output += pool.join('') + '\n'
        output += " ".join(map(str, pool)) + "\n"
        return
    for i in range(1, N+1):
        if not state[i]:
            state[i] = True
            pool.append(i)
            search(deps + 1)
            pool.pop()
            state[i] = False

search(0)
print(output)
```

```python
N, M = map(int, input().split())
ans = []
def combination(arr, index, temp):
    if len(temp) == M:
        ans.append(temp[:])
        return
    for i in range(index, len(arr)):
        # if arr[i] == arr[i-1]: # 이전에 선택한 값과 겹치는 경우 skip
        # if i > index and arr[i] == arr[i-1]: # 이전에 선택한 값과 겹치는 경우 skip
        #     continue
        temp.append(arr[i])
        combination(arr, i, temp)
        temp.pop()
    return ans

print('\n'.join([' '.join(map(str, comb)) for comb in combination([x for x in range(1, N+1)], 0, [])]))
```

```python
n,m = map(int, input().split())
def suyeol(candidate, left,  munja = ''):
    if left == 1:
        for i in candidate:
            print(munja+str(i))
    else:
        for i in candidate:
            newmunja = munja + str(i) + ' ' 
            tmp = candidate[:]
            tmp.remove(i)
            suyeol(tmp, left-1, newmunja)

suyeol([i for i in range(1, n+1)], m)
```
